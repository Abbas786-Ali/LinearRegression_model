```python
#impot the basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#impot the linear regretion model from skitet learn library
from sklearn.linear_model import LinearRegression
```

```python
#uplead a data_set from seabron
df=sns.load_dataset('tips')
df.head()
```

```python
#create a regretions plot
X=df[['total_bill']]
Y=df['tip']
sns.regplot(x=X,y=Y)
```

```python
#now train the model
model=LinearRegression()
model.fit(X,Y)
```

```python
import pandas as pd

# predict the tips for unknow totl bill
model.predict(pd.DataFrame([[50]], columns=['total_bill']))
```
